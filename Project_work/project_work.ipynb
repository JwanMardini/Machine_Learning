# Dataset: Dota2 Games Results

Dota 2 is a popular computer game with two teams of 5 players. At the start of the game each player chooses a unique hero with different strengths and weaknesses. The dataset is reasonably sparse as only 10 of 113 possible heroes are chosen in a given game. All games were played in a space of 2 hours on the 13th of August, 2016

Each row of the dataset is a single game with the following features (in the order in the vector):
1. Team won the game (1 or -1)
2. Cluster ID (related to location)
3. Game mode (eg All Pick)
4. Game type (eg. Ranked)
5. - end: Each element is an indicator for a hero. Value of 1 indicates that a player from team '1' played as that hero and '-1' for the other team. Hero can be selected by only one player each game. This means that each row has five '1' and five '-1' values.

### Merging and Shuffling of Datasets

After mapping the identifiers to readable names, I merged the training and test datasets and shuffled the combined dataset. This process ensures a balanced distribution of data points, reducing biases and enhancing the reliability of the model evaluation.

```python
def merge_and_shuffle(df1, df):
    merged_df = pd.concat([df1, df], ignore_index=True)
    shuffled_df = merged_df.sample(frac=1).reset_index(drop=True)
    return shuffled_df
```

#### Encoding

We applied one-hot encoding to categorical columns in the dataset to prepare them for machine learning models. This encoding method transforms categorical variables into a binary matrix, allowing each unique category to have its own column with a binary indicator.

The function `one_hot_encode_columns()` was used to encode the specified columns (`"cluster_id"`, `"game_mode"`, `"game_type"`) using `OneHotEncoder` from `sklearn.preprocessing`. This function:
1. Encodes the specified columns and creates a new DataFrame with the encoded columns.
2. Concatenates the new columns with the original DataFrame.
3. Removes the original categorical columns.

```python
from sklearn.preprocessing import OneHotEncoder
import pickle

def one_hot_encode_columns(df, columns):
    encoder = OneHotEncoder(sparse=False)
    encoded_data = encoder.fit_transform(df[columns])
    encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(columns))
    df = pd.concat([df.reset_index(drop=True), encoded_df.reset_index(drop=True)], axis=1)
    df.drop(columns, axis=1, inplace=True)
    
    # Save the encoder for future use (for consistency in predictions)
    with open('encoder.sav', 'wb') as f:
        pickle.dump(encoder, f)
    return df
```

```python
from sklearn.preprocessing import LabelEncoder

def label_encode_columns(df, columns):
    """Applies label encoding to specified columns of a DataFrame."""
    df_encoded = df.copy()
    for col in columns:
        le = LabelEncoder()
        df_encoded[col] = le.fit_transform(df_encoded[col])
    return df_encoded
```

```python
df = one_hot_encode_columns(df, ["cluster_id",	"game_mode",	"game_type"])
```

#### Model Evaluation and Hyperparameter Tuning

These functions assist in evaluating and optimizing machine learning models. The `evaluate_model` function provides classification metrics, including accuracy, precision, recall, and the confusion matrix. The `cross_validate_model` function performs k-fold cross-validation to assess model consistency. Finally, `find_best_hyperparameters` uses grid search to fine-tune model hyperparameters and select the best configuration based on accuracy.

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

def evaluate_model(y_test, y_pred):
    """Evaluates a classification model's performance by printing the classification report, confusion matrix, and accuracy score."""
    # Evaluate the model
    print(classification_report(y_test, y_pred))
    print(confusion_matrix(y_test, y_pred))
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy: {accuracy}")
```

```python
from sklearn.model_selection import cross_validate

def cross_validate_model(model, X, y, cv=5):
    """Performs cross-validation on a given model and returns accuracy scores across folds."""
    scores = cross_validate(model, X, y, cv=cv, scoring='accuracy')
    return scores
```

```python
from sklearn.model_selection import GridSearchCV

def find_best_hyperparameters(model, param_grid, X_train, y_train):
    """Uses grid search with cross-validation to find and return the best hyperparameters for a given model."""
    grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='accuracy') # Use 5-fold cross-validation
    grid_search.fit(X_train, y_train)
    print("Best hyperparameters:", grid_search.best_params_)
    print("Best accuracy score:", grid_search.best_score_)
    return grid_search.best_estimator_
```

#### Feature Importence

```python
import numpy as np

def show_feature_importance(model, X, feature_names=None, use_permutation=False):
    """Displays feature importance for a fitted model."""
    if feature_names is None:
        feature_names = [f"Feature {i}" for i in range(X.shape[1])]

    # Check if model has feature importances or coefficients
    if hasattr(model, "feature_importances_"):
        # Tree-based model feature importance
        importances = model.feature_importances_
    elif hasattr(model, "coef_"):
        # For linear models (like Logistic Regression), use absolute values of coefficients
        importances = np.abs(model.coef_[0])
    elif use_permutation:
        # For models without direct importance metrics, use permutation importance
        results = permutation_importance(model, X, model.predict, n_repeats=10, random_state=42)
        importances = results.importances_mean
    else:
        raise ValueError("Model does not support direct feature importance, try enabling use_permutation.")

    # Sort features by importance
    indices = np.argsort(importances)[::-1]
    sorted_features = np.array(feature_names)[indices]
    sorted_importances = importances[indices]

    # Plot feature importance
    plt.figure(figsize=(16, 20))
    plt.barh(sorted_features, sorted_importances, color='skyblue')
    plt.xlabel("Feature Importance")
    plt.title("Feature Importance Ranking")
    plt.gca().invert_yaxis()
    plt.show()
```
